* Enterprise and Midmarket have relatively low and fluctuating profit trends, often under $15K.

**Key Observations**

* Small Business is the top performer in both sales and profit, indicating a strong and consistent contribution to overall revenue.

* Seasonal peaks (especially in July and April for sales, and February for profits) suggest possible promotional or market-cycle effects.

* Channel Partners, despite low sales, achieve relatively better profit margins compared to Midmarket.

* Midmarket underperforms across both metrics, indicating a potential area for strategic review or market focus shift.

**Effect of Discounts on Sales and Profits**

**1. Effects of Discounts on sales**

```python
# efect of discounts on profits
sns.set(style="whitegrid")
sns.lineplot(x='Discount Band', y='Sales', data=mydata, alpha=0.7)

plt.title('Sales vs Discount by Segment')
plt.ylabel('Sales')
plt.xlabel('Discount')
plt.legend(title='Segment')
plt.show()
```

**2. Effects of Discounts on Profits**

```python
# efect of discounts on profits
sns.set(style="whitegrid")
sns.lineplot(x='Discount Band', y='Profit', data=mydata, alpha=0.7)

plt.title('Profit vs Discount by Segment')
plt.ylabel('Profit')
plt.xlabel('Discount')
plt.legend(title='Segment')
plt.show()
```

**Insight: Impact of Discount Levels on Sales and Profit**

* Low discount generated the highest sales ($0.22M) and highest profits ($0.04M).

* No discount achieved decent sales ($0.15M) and profit ($0.033M), showing customers buy even without markdowns.

* Medium discount reduced both sales ($0.16M) and profit ($0.024M).

* High discount had the weakest performance, with sales ($0.15M) and profits ($0.018M).

* Findings suggest low discounts maximize returns, while heavy discounting harms both revenue and profit.

**Cost of Good Sold (COGS) vs Profits**

```python
sns.regplot(x='COGS', y='Profit', data=mydata, scatter_kws={'alpha':0.7})
plt.title('COG vs Profit by Segment')
plt.xlabel('Cost of Goods Sold (COG)')
plt.ylabel('Profit')
plt.legend(title='Segment')
plt.show()
```

**Analysis 5  COG vs Profit Relationship**
* Positive Correlation:

1. As Cost of Goods Sold (COG) increases, Profit also tends to increase.

2. The trendline slope confirms a strong positive relationship.

* Profit Potential:

1. Highest profits (~260,000) occur when COG is in the range of 700,000900,000.

2. Low COG values (<50,000) generally lead to profits below 20,000.

* Outliers:

1. There are points with high profit despite relatively lower COG (under 200,000), which could indicate highly efficient sales or premium pricing.

2. Conversely, some points with high COG (~700,000) still show comparatively lower profits (~50,000), suggesting inefficiency or heavy discounts.

* Risk vs Reward:

1. Higher COG brings higher risk (more capital investment) but usually yields higher profits.

2. Companies operating in the mid-COG range (~200,000500,000) see a wide spread of profits, suggesting other factors like marketing, pricing, and demand fluctuations play a big role.
